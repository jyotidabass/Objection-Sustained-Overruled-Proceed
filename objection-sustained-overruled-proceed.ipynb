When an objection is overruled it means that the evidence is properly admitted to the court, and the trial can proceed. When an objection is sustained, the lawyer must rephrase the question or otherwise address the issue with the evidence to ensure that the jury only hears properly admitted evidence.

https://www.findlaw.com/litigation/going-to-court/how-does-a-judge-rule-on-objections.html#:~:text=When%20an%20objection%20is%20overruled,only%20hears%20properly%20admitted%20evidence.

#Remove Nan to plot the chart below avoiding Nan error.

```python
# categorical features with missing values
categorical_nan = [feature for feature in df.columns if df[feature].isna().sum()>0 and df[feature].dtypes=='O']
print(categorical_nan)
```

```python
# replacing missing values in categorical features
for feature in categorical_nan:
    df[feature] = df[feature].fillna('None')
```

#Checking if the Nans are gone.

```python
fig = px.bar(df[['issue_area','decision_type','first_party_winner']].sort_values('decision_type', ascending=False), 
                        y = "decision_type", x= "issue_area", color='first_party_winner', template='ggplot2')
fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_layout(title_text="Issue Areas Decisions")

fig.show()
```
